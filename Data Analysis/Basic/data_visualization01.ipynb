## 05. 다중 데이터
### COVID-19 백신 종류별 접종 인구 (가상)

```python
days = [1,2,3] # 10월 1일, 2일, 3일 의미
az = [2,4,8] # 단위 : 1만명
pfizer = [5,1,3]
moderna = [1,2,5]

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label = 'AZ')
plt.plot(days, pfizer, label = 'Pfizer', marker = 'o', ls = '--')
plt.plot(days, moderna, label = 'Moderna', marker = 's', ls = '-.')

plt.legend()
```

```python
plt.plot(days, az, label = 'AZ')
plt.plot(days, pfizer, label = 'Pfizer', marker = 'o', ls = '--')
plt.plot(days, moderna, label = 'Moderna', marker = 's', ls = '-.')

plt.legend(ncol=3) #범례표 컬럼 갯수 변경
```

## 09. 막대그래프(심화)

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184]

plt.barh(labels, values)
plt.xlim(175,195)
plt.xticks(rotation = 45)
```

```python
bar = plt.bar(labels,values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
bar = plt.bar(labels, values)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+0.5, values[idx], ha = 'center')
```

## 11. 누적 막대 그래프

```python
plt.bar(df['이름'],df['국어'])
```

```python
plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어']) # 국어 막대 그래프 위에 영어 그래프를 쌓아올린다.
plt.xticks(rotation = 45)
```

```python
plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어'])
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'])
plt.xticks(rotation = 45)
```

```python
plt.bar(df['이름'],df['국어'], label = '국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label = '영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'], label = '수학')
plt.xticks(rotation = 45)
plt.legend()
```

## 12. 다중 막대 그래프

```python
# 기본 원리
index = np.arange(df.shape[0])

w = 0.25
plt.bar(index-w, df['국어'])
```

```python
w = 0.25
plt.bar(index-w, df['국어'], width=w)
plt.bar(index, df['영어'], width=w)
plt.bar(index+w, df['수학'], width=w)
```

```python
plt.figure(figsize=(10,5))
plt.title('학생별 성적')

w = 0.25
plt.bar(index-w, df['국어'], width=w, label = '국어')
plt.bar(index, df['영어'], width=w, label = '영어')
plt.bar(index+w, df['수학'], width=w, label = '수학')
plt.legend(ncol = 3)
plt.xticks(index, df['이름'], rotation = 45)

plt.show()
```
