##### 결측치 제외하기 dropna()
    - Parameter
    (1) axis = index or columns : 결측치가 존재하는 데이터를 삭제할 때, 행을 삭제할 것인지 열을 삭제할 것인지 선택
    (2) how = any or all : 결측치가 행 또는 열에서 하나라도 발견되면 삭제할지, 데이터 행 또는 열 전체가 결측치일 때 삭제할 지 선택 (any가 default)

```python
df.dropna()
```

```python
df.dropna(axis= 'index', how='any') # NaN이 하나라도 있는 row 삭제
df.dropna(axis= 'columns') # NaN이 하나라도 있는 column 삭제 how = 'any' 기본 설정
df.dropna(axis= 'index', how='all') # NaN이 모두 있는 row 삭제

df['학교'] = np.nan
df
df.dropna(axis = 'columns', how = 'all') # 전부 NaN이었던 학교 column이 삭제됨
```

##### 데이터 수정 & 데이터 함수 적용 (apply)

```python
df1 = df.copy(True)
```

```python
# 키 데이터 뒤에 'cm'를 추가하려한다.
# -> int + str로 인해 error가 발생한다.
# => int를 str로 바꿔주는 함수를 거친 뒤 'cm'을 붙인다.
```

```python
def add_cm(height):
    return str(height) + 'cm'

df1['키'] = df1['키'].apply(add_cm)
df1
```

```python
# SW특기 데이터들을 모두 첫글자는 대문자, 나머지는 소문자로 변환하자
def capitalize(lang):
    if pd.notna(lang):
        return lang.capitalize()
    return lang

df1['SW특기'] = df1['SW특기'].apply(capitalize)
```
