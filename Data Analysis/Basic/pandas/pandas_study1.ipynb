```python
df1 = pd.DataFrame(df.groupby(['pclass', 'survived'])['name'].count())
```

```python
df2 = df1.reset_index()
df2
```

```python
df2.groupby('pclass')['name'].sum()
```

```python
port_list = []
for i in df2.index:
    portion = round(df2.iloc[i, 2] / df2[df2['pclass'] == df2.iloc[i,0]]['name'].sum(),2)
    port_list.append(portion)

df2['portion'] = port_list
df2

# 탑승 등급별로 사망한 비율이 다르다.
# 1등급 승객 중 사망자는 38%
# 2등급 승객 중 사망자는 57%
# 3등급 승객 중 사망자는 74%
```
